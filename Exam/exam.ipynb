`[0.2]` Create a new column `whoWon` which contains `"red"` or `"blue"` based on the `blueWins` column.

````python
def who(row):
    if(row.blueWins==1):
        return "blue"
    else:
        return "red"

whoWon=lol_data.apply(who,axis="columns")
lol_data['whoWon']=whoWon
lol_data.head(15)
````

`[0.2]` Create a new column `firstKill` which contains `"red"` or `"blue"` based on either `blueFirstBlood` or `redFirstBlood`.

````python
def who(row):
    if(row.redFirstBlood==1):
        return "blue"
    else:
        return "red"

firstKill=lol_data.apply(who,axis="columns")
lol_data['firstKill']=firstKill
lol_data.head(15)
````
